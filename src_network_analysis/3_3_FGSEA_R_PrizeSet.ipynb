```r
# gene set combined
output_dir = '../GSEA/fgsea/gene_set_B50_G5/'

pathways_prize <- gmtPathways(paste(output_dir, "gene_set_prize_combined.gmt", sep=""))
str(head(pathways_prize))
# clean up by removing empty strings
# access actual values by [[]]
for (gene_set in names(pathways_prize)) {
    print(gene_set)
    pathways_prize[[gene_set]] <- pathways_prize[[gene_set]][nchar(pathways_prize[[gene_set]]) != 0]
}
str(pathways_prize)
```

```r
# test plot
# adult data Ex fly
# make list of ranks
output_dir = "../GSEA/fgsea/"
cell_types = c('Ex','In','Ast','Oli','Opc','Mic')
ranks_adult = list()
for (cell in cell_types) {
    print(cell)
    filename = paste(output_dir, cell, "_wilcoxon_0.05_adult_cortex.rnk", sep="")
    print(filename)
    ranks_adult[[cell]] = read.table(filename,
                    header=FALSE, colClasses = c("character", "numeric"))
    ranks_adult[[cell]] <- setNames(ranks_adult[[cell]]$V2, ranks_adult[[cell]]$V1) # set names (gene ID) to rank metric
}

```

```r
pathway<-pathways_prize[['Fly']]
cell <- 'Ex'
#plot <- plotEnrichment(pathway, ranks_adult[[cell]])
plot <- plotEnrichment(pathway, ranks_adult[[cell]]) +
  theme_minimal(base_family = "Arial") +  # Set the font to Arial
  theme(
    axis.title = element_text(size = 30, face = "bold"),  # Increase axis title font size
    axis.text = element_text(size = 18),  # Increase axis tick label font size
    panel.grid.major = element_blank(),  # Remove major grid lines
    panel.grid.minor = element_blank(),  # Remove minor grid lines
    axis.line = element_line(size = 1.2)  # Make x and y axis lines thicker
    #axis.line = element_line(colour = "black", size = 1.2)
  ) +
  labs(x = "Rank in Ordered Dataset", y = "Enrichment Score")
```

```r
print(plot)
```

```r
# Save the plot as a .png file
ggsave("../Figures/adult_Ex_Fly_Prize_enrichment_plot.png", plot = plot, width = 10, height = 7)

# Save the plot as a .eps file
ggsave("../Figures/adult_Ex_Fly_Prize_enrichment_plot.eps", plot = plot, width = 10, height = 7, device = "eps")
```

## This notebook runs FGSEA for assessing cell type specificity and makes enrichment plot

```r
library(clusterProfiler); library(msigdbr); library(fgsea)
library(data.table)
library(ggplot2)
library(dplyr)
```

```r
 require(ggplot2); require(tidyr); require(Biobase);require('tidyverse')
```

## Combined Prize gene set

```r
# gene set combined
output_dir = '../GSEA/fgsea/gene_set_B50_G5/'

pathways_prize <- gmtPathways(paste(output_dir, "gene_set_prize_combined.gmt", sep=""))
str(head(pathways_prize))
# clean up by removing empty strings
# access actual values by [[]]
for (gene_set in names(pathways_prize)) {
    print(gene_set)
    pathways_prize[[gene_set]] <- pathways_prize[[gene_set]][nchar(pathways_prize[[gene_set]]) != 0]
}
str(pathways_prize)
```

Adult data FGSEA

```r
# make list of ranks
output_dir = "../GSEA/fgsea/"
cell_types = c('Ex','In','Ast','Oli','Opc','Mic')
ranks_adult = list()
for (cell in cell_types) {
    print(cell)
    filename = paste(output_dir, cell, "_wilcoxon_0.05_adult_cortex.rnk", sep="")
    print(filename)
    ranks_adult[[cell]] = read.table(filename,
                    header=FALSE, colClasses = c("character", "numeric"))
    ranks_adult[[cell]] <- setNames(ranks_adult[[cell]]$V2, ranks_adult[[cell]]$V1) # set names (gene ID) to rank metric
}

# fgsea 
# adult rank vs gene set combined
fgseaRes_adult_prize = list()
cell_types = c('Ex','In','Ast','Oli','Opc','Mic')
for (cell in cell_types) {
    print(cell)
    fgseaRes_adult_prize[[cell]] <- fgsea(pathways_prize, ranks_adult[[cell]], nPermSimple = 100000)
}
fgseaRes_adult_prize
```

```r
# make list of ranks
output_dir = "../GSEA/fgsea/"
cell_types = c('Ex','In','Ast','Oli','Opc','Mic')
ranks_juv = list()
for (cell in cell_types) {
    print(cell)
    filename = paste(output_dir, cell, "_wilcoxon_0.05_juv_cortex.rnk", sep="")
    print(filename)
    ranks_juv[[cell]] = read.table(filename,
                    header=FALSE, colClasses = c("character", "numeric"))
    ranks_juv[[cell]] <- setNames(ranks_juv[[cell]]$V2, ranks_juv[[cell]]$V1) # set names (gene ID) to rank metric
}


# fgsea 
# juv rank vs prize
fgseaRes_juv_prize = list()
cell_types = c('Ex','In','Ast','Oli','Opc','Mic')
for (cell in cell_types) {
    print(cell)
    fgseaRes_juv_prize[[cell]] <- fgsea(pathways_prize, ranks_juv[[cell]], nPermSimple = 100000)
}
fgseaRes_juv_prize
```

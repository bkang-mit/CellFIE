```python
# generate list of genes by merging overlapping genes.
GEM_eQTL_meta_list = np.unique(list(GEM_eQTL_meta_df['eQTL Gene']))
VZ_eQTL_meta_list = np.unique(list(VZ_eQTL_meta_df['eQTL_gene']))
print(len(GEM_eQTL_meta_list))
print(len(VZ_eQTL_meta_list))
```

Annotate effect size and GWA pvalue. Note that Venezuelan data were processed with "Yeo Johnson" Power transformation on the residual age of onset and TopMed population imputation of the genotypes, so the effect size range is different from that of GeMHD-Mega

For each gene, select SNP-gene pair with max effect size and its associated pvalue, to assign effect size / pval

```python
# process positive and negative modifiers separately. 
# GEMHD Mega
pos_GEM_eqtl_df = GEM_eQTL_meta_df[GEM_eQTL_meta_df['Est_assoc'] > 0 ]
neg_GEM_eqtl_df = GEM_eQTL_meta_df[GEM_eQTL_meta_df['Est_assoc'] <= 0 ]

# VZ
pos_VZ_eqtl_df = VZ_eQTL_meta_df[VZ_eQTL_meta_df['beta'] > 0 ]
neg_VZ_eqtl_df = VZ_eQTL_meta_df[VZ_eQTL_meta_df['beta'] <= 0 ]

# GEMHD Mega
pos_GEM_eQTL_meta_list = np.unique(list(pos_GEM_eqtl_df['eQTL Gene']))
neg_GEM_eQTL_meta_list = np.unique(list(neg_GEM_eqtl_df['eQTL Gene']))

# VZ
pos_VZ_eQTL_meta_list = np.unique(list(pos_VZ_eqtl_df['eQTL_gene']))
neg_VZ_eQTL_meta_list = np.unique(list(neg_VZ_eqtl_df['eQTL_gene']))
```

```python
# select maximum effect size snp for each eqtl hit gene
GEM_effect_sizes = {}
GEM_pvals= {}
both_effect_list=[]
# first assign positive modifier effect size.
for gene in pos_GEM_eQTL_meta_list:
    df = pos_GEM_eqtl_df[pos_GEM_eqtl_df['eQTL Gene'] == gene]

    GEM_effect_sizes[gene] = abs(df['Est_assoc']).max()
    
    GEM_pvals[gene] = df[df['Est_assoc'] == GEM_effect_sizes[gene]]['Wald.pval_assoc'].values[0]
# assign negative modifier effect size. If a gene is annotated as both pos/negative, then assign larger effect size.    
for gene in neg_GEM_eQTL_meta_list:
    df = neg_GEM_eqtl_df[neg_GEM_eqtl_df['eQTL Gene'] == gene]
    if gene in GEM_effect_sizes.keys(): # if gene is in positive, add it to 'both'
        both_effect_list.append(gene)
        if GEM_effect_sizes[gene] > 0:
            if GEM_effect_sizes[gene] > abs(df['Est_assoc'].min()):
                # if positive effect size is greater than negative, skip
                pass
            else:
                # if positive effect size is lower, update with negative 
                GEM_effect_sizes[gene] = df['Est_assoc'].min()
                GEM_pvals[gene] = df[df['Est_assoc'] == GEM_effect_sizes[gene]]['Wald.pval_assoc'].values[0]
    else:
        GEM_effect_sizes[gene] = df['Est_assoc'].min()
        GEM_pvals[gene] = df[df['Est_assoc'] == GEM_effect_sizes[gene]]['Wald.pval_assoc'].values[0]
        
# test run to verify. UROS has both positive / negative effect size.
print(GEM_effect_sizes['UROS'])
df2 = neg_GEM_eqtl_df[neg_GEM_eqtl_df['eQTL Gene'] == 'UROS']
df3 = pos_GEM_eqtl_df[pos_GEM_eqtl_df['eQTL Gene'] == 'UROS']

print(df2['Est_assoc'].min())
print(df3['Est_assoc'].max())
```

```python
# inspect "both" effect sizes;
both_effect_list
```

```python
# apply same processing for VZ data
VZ_effect_sizes = {}
VZ_pvals={}
for gene in pos_VZ_eQTL_meta_list:
    df = pos_VZ_eqtl_df[pos_VZ_eqtl_df['eQTL_gene'] == gene]

    VZ_effect_sizes[gene] = abs(df['beta']).max()
    VZ_pvals[gene] = df[df['beta'] == VZ_effect_sizes[gene]]['neg_log_pvalue'].values[0]
    
for gene in neg_VZ_eQTL_meta_list:
    df = neg_VZ_eqtl_df[neg_VZ_eqtl_df['eQTL_gene'] == gene]
    if gene in VZ_effect_sizes.keys():
        
        both_effect_list.append(gene)
        if VZ_effect_sizes[gene] > 0:
            if VZ_effect_sizes[gene] > abs(df['beta'].min()):
                
                pass
            else:
                VZ_effect_sizes[gene] = df['beta'].min()
                VZ_pvals[gene] = df[df['beta'] == VZ_effect_sizes[gene]]['neg_log_pvalue'].values[0]
    else:
        VZ_effect_sizes[gene] = df['beta'].min()
        VZ_pvals[gene] = df[df['beta'] == VZ_effect_sizes[gene]]['neg_log_pvalue'].values.max()
print(VZ_effect_sizes)
```

```python
# Checking min/max effect sizes of two sources
# GEM Mega
print('GeMHD-Mega Analysis')
pos_GEM_meta_df = GEM_meta_df[GEM_meta_df['Est_assoc'] > 0]
neg_GEM_meta_df = GEM_meta_df[GEM_meta_df['Est_assoc'] <= 0]

pos_GEM_max = pos_GEM_meta_df['Est_assoc'].max()
pos_GEM_min = pos_GEM_eqtl_df['Est_assoc'].min() # using data from eQTL list for true minimum..
neg_GEM_max = neg_GEM_meta_df['Est_assoc'].min()
neg_GEM_min = neg_GEM_eqtl_df['Est_assoc'].max() # using data from eQTL list for true minimum..
print(pos_GEM_max, pos_GEM_min, neg_GEM_max, neg_GEM_min)

# VZ
print('VZ Analysis')
pos_VZ_meta_df = VZ_meta_df[VZ_meta_df['beta']>0]
neg_VZ_meta_df = VZ_meta_df[VZ_meta_df['beta']<=0]

pos_VZ_max = pos_VZ_meta_df['beta'].max()
pos_VZ_min = pos_VZ_meta_df['beta'].min()
neg_VZ_max = neg_VZ_meta_df['beta'].min()
neg_VZ_min = neg_VZ_meta_df['beta'].max()
print(pos_VZ_max, pos_VZ_min, neg_VZ_max, neg_VZ_min)
```

```python
# generate effect size dataframe
effect_size_df = pd.DataFrame(index=final_list, columns=['Direction','effect_size', 'p-val',
                                                         '-log(p)','Z-score','Z-score_pval', 'data_source'])

# for each row, annotate each column;
# Direction: "Positive" means later age of onset, "Negative" means earlier age of onset. Signs of effect size
# effect_size: effect size before normalization
# data_source: source of the data and type of fine-mapping (eQTL vs coding, GeM Mega vs VZ)
# p-val: pval from GWA
# Z-score: Z score of effect size, accounting for different transformations in two studies
# Z-score_pval: Z score of p-val

for index, row in effect_size_df.iterrows():
    if index in GEM_effect_sizes.keys():
        # eQTL GEM
        if GEM_effect_sizes[index] > 0: # positive effect size
            row['Direction']='Positive'
            row['effect_size'] = GEM_effect_sizes[index]
            row['data_source']='GEM Mega eQTL'
            row['p-val'] = GEM_pvals[index]
            row['-log(p)'] = -np.log(GEM_pvals[index])
        else: # negative
            row['Direction']='Negative'
            row['effect_size'] = GEM_effect_sizes[index]
            row['data_source']='GEM Mega eQTL'
            row['p-val'] = GEM_pvals[index]
            row['-log(p)'] = -np.log(GEM_pvals[index])
            
            
    elif index in VZ_effect_sizes.keys():
        # VZ eQTL 
        if VZ_effect_sizes[index] > 0:

            row['Direction']='Positive'
            row['effect_size'] = VZ_effect_sizes[index]
            row['data_source']='VZ eQTL'
            row['p-val'] = VZ_pvals[index]
            row['-log(p)'] = VZ_pvals[index] # it was already -log transformed
            
        else: # negative
            row['Direction']='Negative'
            row['effect_size'] = VZ_effect_sizes[index]
            row['data_source']='VZ eQTL'
            row['p-val'] = VZ_pvals[index]
            row['-log(p)'] = VZ_pvals[index]
            
    # GEM coding variants        
    elif index in GEM_coding_effect_sizes.keys():
        if GEM_coding_effect_sizes[index] > 0:
            
            row['effect_size'] =GEM_coding_effect_sizes[index]
            row['Direction']='Positive'
            row['data_source']='GEM Mega missense'
            row['p-val'] = GEM_coding_pvals[index]
            row['-log(p)'] = -np.log(GEM_coding_pvals[index])
        else:
            row['effect_size'] =GEM_coding_effect_sizes[index]
            row['Direction']='Negative'
            row['data_source']='GEM Mega missense'
            row['p-val'] = GEM_coding_pvals[index]
            row['-log(p)'] = -np.log(GEM_coding_pvals[index])
    # VZ coding variants
    elif index in VZ_coding_effect_sizes.keys():
        if VZ_coding_effect_sizes[index] > 0:

            row['effect_size'] =VZ_coding_effect_sizes[index]
            row['Direction']='Positive'
            row['data_source']='VZ missense'
            row['p-val'] = VZ_coding_pvals[index]
            row['-log(p)'] = VZ_coding_pvals[index]
        else:
            row['effect_size'] =VZ_coding_effect_sizes[index]
            row['Direction']='Negative'
            row['data_source']='VZ missense'
            row['p-val'] = VZ_coding_pvals[index]
            row['-log(p)'] = VZ_coding_pvals[index]
            
for index, row in effect_size_df.iterrows():    
    if index in both_effect_list:
        row['Direction']='Both'
```

```python
# annotate z-score
GEM_mean_effect_size = abs(GEM_meta_df['Est_assoc']).mean()
GEM_std_effect_size = abs(GEM_meta_df['Est_assoc']).std(ddof=0)
GEM_mean_pval = (-np.log(GEM_meta_df['Wald.pval_assoc'])).mean()
GEM_std_pval = (-np.log(GEM_meta_df['Wald.pval_assoc'])).std(ddof=0)

VZ_mean_effect_size = abs(VZ_meta_df['beta']).mean()
VZ_std_effect_size = abs(VZ_meta_df['beta']).std(ddof=0)
VZ_mean_pval = VZ_meta_df['neg_log_pvalue'].mean()
VZ_std_pval = VZ_meta_df['neg_log_pvalue'].std(ddof=0)

# annotate p-val z score
zscores = {}
for index, row in effect_size_df.iterrows():
    if row['data_source'] == 'GEM Mega eQTL':
        row['Z-score'] = (abs(row['effect_size']) - GEM_mean_effect_size)/GEM_std_effect_size
    elif row['data_source'] == 'GEM Mega missense':
        row['Z-score'] = (abs(row['effect_size']) - GEM_mean_effect_size)/GEM_std_effect_size
    elif row['data_source'] == 'VZ eQTL':
        row['Z-score'] = (abs(row['effect_size']) - VZ_mean_effect_size)/VZ_std_effect_size
    elif row['data_source'] == 'VZ missense':
        row['Z-score'] = (abs(row['effect_size']) - VZ_mean_effect_size)/VZ_std_effect_size
    zscores[index] =row['Z-score']

print(GEM_mean_pval,GEM_std_pval,VZ_mean_pval,VZ_std_pval)
z_pvals = {}
for index, row in effect_size_df.iterrows():
    if row['data_source'] == 'GEM Mega eQTL':
        row['Z-score_pval'] = (abs(row['-log(p)']) - GEM_mean_pval)/GEM_std_pval
    elif row['data_source'] == 'GEM Mega missense':
        row['Z-score_pval'] = (abs(row['-log(p)']) - GEM_mean_pval)/GEM_std_pval
    elif row['data_source'] == 'VZ eQTL':
        row['Z-score_pval'] = (abs(row['-log(p)']) - VZ_mean_pval)/VZ_std_pval
    elif row['data_source'] == 'VZ missense':
        row['Z-score_pval'] = (abs(row['-log(p)']) - VZ_mean_pval)/VZ_std_pval
    z_pvals[index] = row['Z-score_pval']
```

```python
# finally, calculate prize value by Z-score + Z-score_pval.
# This is a score that reflects 1) relative confidence, 2) relative effect sizes within each dataset.
# offset will be made in final prize value
effect_size_df['prize'] = effect_size_df['Z-score'] + effect_size_df['Z-score_pval']
effect_size_df
```

Update prize based on each cell type specificity, using preprocessed snRNA seq data (Lim et al 2022). We used prepublished version (see also prize_update 100621 in cluster)

```python
import pandas as pd
import scanpy as sc
import os
from scipy.io import mmread
from scipy.sparse import csr_matrix
import pickle
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import iqr
import seaborn as sns
```

```python
# remove 'N/A' cells - these cells are stored with the data, so repeat this process whenever loading adata
adult_cortex = adult_cortex[adult_cortex.obs['annot'] != 'N/A']
juv_cortex = juv_cortex[juv_cortex.obs['annot'] != 'N/A']
print(adult_cortex.shape, juv_cortex.shape)
```

```python
# Load prize (scaled to 0 and 1)
prize_df = pd.read_csv('../Prize_Inputs/prize_scaled.txt', header = 0, index_col = 0, delimiter = '\t')
prize_df
```

```python
# remove duplicate index
prize_df = prize_df.loc[~prize_df.index.duplicated(), :]
prize_df
```

```python
# defining mean expression profiles for each cell type.
def get_mean_exp(adatas, prize_df, celltypes):
    adult_genes = []
    juv_genes= []
    # dealing with genes present in prize_df AND adata
    for gene in prize_df.index:
        if gene in adatas[0].var_names:
            adult_genes.append(gene)
    
    for gene in prize_df.index:
        if gene in adatas[1].var_names:
            juv_genes.append(gene)

    # create dataframe with only genes that are found in prize list
    mean_adult = pd.DataFrame(index = adult_genes, columns = ['mean_exp', 'STD'])
    mean_juv = pd.DataFrame(index = juv_genes, columns = ['mean_exp', 'STD'])
    
    adata_adult = adatas[0]
    adata_juv = adatas[1]
    # all cells
    mean_adult['mean_exp'] = np.mean(adata_adult[:, adult_genes].X.toarray(), axis=0)
    mean_juv['mean_exp'] = np.mean(adata_juv[:, juv_genes].X.toarray(), axis=0)                                                   
    mean_adult['STD'] =  np.std(adata_adult[:,adult_genes].X.toarray(), axis = 0)
    mean_juv['STD'] = np.std(adata_juv[:,juv_genes].X.toarray(), axis = 0)
    # cell median
    for celltype in celltypes:
        mean_adult[celltype] = np.mean(adata_adult[adata_adult.obs['annot'] == celltype, adult_genes].X.toarray(), axis=0)
        mean_juv[celltype] = np.mean(adata_juv[adata_juv.obs['annot'] == celltype, juv_genes].X.toarray(), axis=0) 
    # create final output dataframe, where all of prize genes are used
    mean_adult_df = pd.DataFrame(index=  prize_df.index.tolist(), columns = ['mean_exp','STD'] + celltypes)
    mean_juv_df = pd.DataFrame(index=  prize_df.index.tolist(), columns = ['mean_exp','STD'] + celltypes)
    
    for gene in mean_adult.index.tolist():
        mean_adult_df.loc[gene, :] = mean_adult.loc[gene, :]
    for gene in mean_juv.index.tolist():
        mean_juv_df.loc[gene, :] = mean_juv.loc[gene, :]
    mean_adult_df.fillna(0, inplace=True)
    mean_juv_df.fillna(0, inplace=True)
    return mean_adult_df, mean_juv_df

# from mean expression profile of each celltype, generate celltype scores. 
# [mean exp(t) - mean exp (all)]/std exp (all0)
def get_cell_score(mean_adult_df, mean_juv_df, celltypes):
    score_df = pd.DataFrame(index = prize_df.index.tolist(), columns = celltypes)
    for celltype in celltypes:
        score_df.loc[:,celltype] = ((mean_adult_df[celltype] - mean_adult_df['mean_exp'])/mean_adult_df['STD'] + 
                                    (mean_juv_df[celltype] - mean_juv_df['mean_exp'])/mean_juv_df['STD'])/2
        
    score_df.fillna(0, inplace=True)
    return score_df
```

```python
# 
adatas = [adult_cortex, juv_cortex]
adult_genes=[]
juv_genes=[]
for gene in prize_df.index:
    if gene in adatas[0].var_names:
        adult_genes.append(gene)
    
for gene in prize_df.index:
    if gene in adatas[1].var_names:
        juv_genes.append(gene)
print("number of genes detected in snRNA seq data")        
print(len(adult_genes), len(juv_genes))
```

```python
# run above function to generate scores
mean_adult, mean_juv = get_mean_exp([adult_cortex, juv_cortex], prize_df, ['Ex','In','Ast','Oli','Opc','Mic'])
score_df=  get_cell_score(mean_adult, mean_juv, ['Ex','In','Ast','Oli','Opc','Mic'])
# max_absolute scaling (normalize by maximum absolute value), to rescale into [-1, 1]
celltypes = ['Ex','In','Ast','Oli','Opc','Mic']
for cell in celltypes:
    score_df[cell] = score_df[cell] / abs(score_df[cell]).max()
score_df
```

```python
# generate adjusted prizes.
# prize = prize + score X weight
# weight is a scaling factor that determines how much we penalize/reward prize based on cell type specificity.
prize_cortex_1 = {}
prize_cortex_5 = {}
prize_cortex_10 = {}
prize_cortex_20 = {}

celltypes = ['Ex','In','Ast','Oli','Opc','Mic']
for celltype in celltypes:
    prize_cortex_1[celltype] = prize_df.copy()
    prize_cortex_5[celltype] = prize_df.copy()
    prize_cortex_10[celltype] = prize_df.copy()
    prize_cortex_20[celltype] = prize_df.copy()
    
for celltype in celltypes:
    # store original prizes
    prize_cortex_1[celltype]['naive_prize'] = prize_cortex_1[celltype]['prize_val'].copy()
    prize_cortex_5[celltype]['naive_prize'] = prize_cortex_5[celltype]['prize_val'].copy()
    prize_cortex_10[celltype]['naive_prize'] = prize_cortex_10[celltype]['prize_val'].copy()
    prize_cortex_20[celltype]['naive_prize'] = prize_cortex_20[celltype]['prize_val'].copy()

    # update prize with prize + s*w
    prize_cortex_1[celltype]['prize_val'] += score_df[celltype]*0.01
    prize_cortex_5[celltype]['prize_val'] += score_df[celltype]*0.05
    prize_cortex_10[celltype]['prize_val'] += score_df[celltype]*0.1
    prize_cortex_20[celltype]['prize_val'] += score_df[celltype]*0.2
    
    
    # add score column
    prize_cortex_1[celltype]['score'] = score_df[celltype]*0.01
    prize_cortex_5[celltype]['score'] = score_df[celltype]*0.05
    prize_cortex_10[celltype]['score'] = score_df[celltype]*0.1
    prize_cortex_20[celltype]['score'] = score_df[celltype]*0.2
    
```

```python
# load interactome (cell type naive), cell type rank extracted from adata (Osama dataset)
int_df = pd.read_csv('../Interactome/iref17_iref14_braInMap_ptau_metab_mapped.txt', sep='\t')
# gene-level annotation exported from adata
adult_cortex_df = pd.read_csv('/nfs/latdata/bkang/osama_scrna/osama_scrna/adata_ver3/adult_cortex_var.csv', index_col=0)
juv_cortex_df = pd.read_csv('/nfs/latdata/bkang/osama_scrna/osama_scrna/adata_ver3/juv_cortex_var.csv', index_col=0)
int_df
```

```python
# take average mean rank percentile across two datasets;
intersection_cortex = list(set(adult_cortex_df.index.tolist()).intersection(juv_cortex_df.index.tolist()))
union_cortex = set(adult_cortex_df.index.tolist() + juv_cortex_df.index.tolist()) 
# # fill in intersection with average of two
cols_cortex = ['Ast_rank', 'Ex_rank', 'In_rank','Opc_rank', 'Oli_rank', 'Mic_rank']
cortex_df = pd.DataFrame(index= union_cortex, columns = cols_cortex)

cortex_df.loc[intersection_cortex] = (adult_cortex_df.loc[intersection_cortex, cols_cortex] + juv_cortex_df.loc[intersection_cortex, cols_cortex])/2

# genes unique to each dataset will be penalized by taking average with empty
adult_unique_cortex = list(set(adult_cortex_df.index.tolist()) - set(juv_cortex_df.index.tolist()))
juv_unique_cortex = list(set(juv_cortex_df.index.tolist()) - set(adult_cortex_df.index.tolist()))

cortex_df.loc[adult_unique_cortex] = adult_cortex_df.loc[adult_unique_cortex, cols_cortex]/2
cortex_df.loc[juv_unique_cortex] = juv_cortex_df.loc[juv_unique_cortex, cols_cortex]/2
```

```python
cortex_df.loc['LRRIQ1']
```

```python
# annotate each protein based on rank percentile.
# Between protein 1 and protein 2, take smaller rank percentile (low exp) 

# initialize with 0's
rank_df1 = int_df.copy()
rank_df1.set_index('protein1', inplace=True)
rank_df1[cortex_df.columns] = 0.0

# fill in actual, observed rank percentiles
# start with getting index that are present in both expr file and int_df
indices_to_update = list(set(cortex_df.index.tolist()).intersection(rank_df1.index.tolist()))
print(len(indices_to_update))
rank_df1.loc[indices_to_update, cortex_df.columns] = cortex_df.loc[indices_to_update, cortex_df.columns]

# rank_df2
rank_df2 = int_df.copy()
rank_df2.set_index('protein2', inplace=True)

# initialize with 0's
rank_df2[cortex_df.columns] = 0.0

# fill in actual, observed rank percentiles
# start with getting index that are present in both expr file and int_df
indices_to_update = list(set(cortex_df.index.tolist()).intersection(rank_df2.index.tolist()))
rank_df2.loc[indices_to_update, cortex_df.columns] = cortex_df.loc[indices_to_update, cortex_df.columns]

# combine two with min_rank, except for the protein-metab interaction case
adj_df = int_df.copy()
temp_df = int_df.copy()
for col in cortex_df.columns:
    temp_df[col+'1'] = rank_df1[col].values
    temp_df[col+'2'] = rank_df2[col].values
# fill in p-m with protein1 rank values
metab_ind = [x for x in int_df.index if int_df.loc[x,  'protein2'][:4] == 'HMDB']
reg_ind = [x for x in int_df.index if int_df.loc[x, 'protein2'][:4] != 'HMDB']

for col in cortex_df.columns:
    adj_df.loc[metab_ind, col] = temp_df.loc[metab_ind, col+'1'].values # p-m gets protein 1 expr score
    adj_df.loc[reg_ind, col] = np.min(temp_df.loc[reg_ind, [col+'1', col+'2']].values, axis=1) # p-p gets min(expr_score)
adj_df
```

```python
DATA_DIRs = ['101222_D10_Coverslip1_Processed', '102022_D10_Coverslip2_Reimage_Processed',
             '101422_D10_Coverslip3_Processed', '102222_D10_Coverslip4_Reimage_Processed',
             '102422_D21_Coverslip7_Processed', '102522_D21_Coverslip8_Processed',
             '102622_D21_Coverslip9_Processed', '102822_D21_Coverslip11_Processed',
             '102922_D21_Coverslip12_Processed', '103122_D28_Coverslip13_Processed',
             '110122_D28_Coverslip14_Processed', '110222_D28_Coverslip15_Processed',
             '110622_D28_Coverslip16_Processed', 
             '110322_D28_Coverslip17_Processed', '110722_D28_Coverslip18_Processed']
META_DIR = 'metadata'
MORPH_DIR = 'morphology_data'
ADATA_DIR = 'adata_Cellpose'
```

```python
# load metadata - load full codebook as well
full_codebook = pd.read_csv(f'../{META_DIR}/full_codebook.csv',sep=',', index_col=0) # this is "legal" codebook
Procode_gRNA = pd.read_csv(f'../{META_DIR}/PROCODE_gRNA.csv',sep=',')
legal_codes = sorted(list(set(Procode_gRNA['ProCode ID'].to_list())))
codebook = full_codebook[legal_codes]
sns.heatmap(codebook, linewidths = 0.3)
markers = pd.read_csv(f'../{META_DIR}/markers.csv')
markers.head(20)
```

```python
# gene to barcode map
## Assign Gene ID
procode_gene_df = pd.read_csv('../metadata/PROCODE_gRNA.csv')
procode_to_gene = {}
for cell in procode_gene_df.index:
    procode = procode_gene_df.loc[cell,'ProCode ID']
    if procode not in procode_to_gene.keys():
        procode_to_gene[procode] = procode_gene_df.loc[cell,'Gene Target']
```

```python
for DATA in DATA_DIRs:
    with open (f'../{DATA}/morphology_data/Neurite_dfs.pkl', 'rb') as handle:
        Neurite_dfs = pickle.load(handle)
    with open(f'../{DATA}/morphology_data/Branch_dfs.pkl','rb') as handle:
        Branch_dfs = pickle.load(handle)
    
    # make column names consistent;
    for key in Neurite_dfs.keys():
        Neurite_dfs[key] = Neurite_dfs[key].rename(procode_to_gene, axis=1)
        Branch_dfs[key] = Branch_dfs[key].rename(procode_to_gene, axis=1)
        
    # Trunks, Sholl and N
    with open(f'../{DATA}/morphology_data/morphology_data.pkl', 'rb') as handle:
        morph_df = pickle.load(handle)
    
    # define allFOVs - all FOV present in dataset
    allFOVs = [x.split('_')[0][1:] for x in set(morph_df.FileName_max)]    
    rep_dfs = {}
    keys = ['N_cell','L','L1','L2','B','B1','B2','T','S','N'] # N_cells are all cells present in FOV, 
    # N is num_cells after filtering for out-of-bound cells (when doing sholl analysis)
    
    for key in keys: # rep_dfs['L'] is image replicate by genes, Length params.
        rep_dfs[key] = pd.DataFrame(index = list(range(9)), columns = [procode_to_gene[c] for c in codebook.columns])
    
    for i in range(9):
        FOVs = [str(x).zfill(3) for x in list(range(225))[0+i*25:25+i*25]]
        
        for gene in rep_dfs['N_cell'].columns: # for each gene,
            N_cell = 0; N_Branches = 0; L = 0; L1 = 0; L2 = 0
            B= 0; B1 = 0; B2 = 0; T = 0; S = 0; N = 0;
            
            for fov in FOVs:
                if fov in allFOVs:
                    fname = f'F{fov}_max.tif'
                    cell_df = morph_df[(morph_df.Gene == gene)&(morph_df.FileName_max==fname)]  
                    N_cell = N_cell + len(cell_df) # sum across all chunks, pre-filter, all cells
                    
                    L = L + Neurite_dfs[(1,1)].loc[fov, gene]
                    B = B + Branch_dfs[(1,1)].loc[fov, gene]
                    L1 = L1 + Neurite_dfs[(3,50)].loc[fov, gene]
                    B1 = B1 + Branch_dfs[(3,50)].loc[fov, gene]
                    L2 = L2 + Neurite_dfs[(5,100)].loc[fov, gene]
                    B2 = B2 + Branch_dfs[(5,100)].loc[fov, gene]
                    
                    cell_df = cell_df[~cell_df[['Trunks','Sholl_1']].isnull().any(axis=1)] # these are edge cases
                    cell_df = cell_df[cell_df.mask_bbox == 0] # exclude black box masked cells.
                    T = T + cell_df.Trunks.sum(axis=0)
                    S = S + cell_df.Sholl_1.sum(axis=0)
                    N = N + len(cell_df)
                rep_dfs['N_cell'].loc[i, gene] = N_cell
                rep_dfs['L'].loc[i, gene] = L
                rep_dfs['L1'].loc[i,gene] = L1
                rep_dfs['L2'].loc[i,gene] = L2
                rep_dfs['B'].loc[i,gene] = B
                rep_dfs['B1'].loc[i,gene] = B1
                rep_dfs['B2'].loc[i,gene] = B2
                
                rep_dfs['T'].loc[i, gene] = T
                rep_dfs['S'].loc[i, gene] = S
                rep_dfs['N'].loc[i, gene] = N

    # post-filtering to remove Field without any cells (image excluded)
    for key in keys:
        df = rep_dfs[key]
        df = df.loc[~(df == 0).all(axis=1)]
        if len(df) != len(rep_dfs[key]):
            print(DATA, key, 'excluding all zero row')
        rep_dfs[key] = df.copy()
        
                
    data_dfs = rep_dfs.copy()
    data_dfs['L'] = rep_dfs['L']/data_dfs['N_cell']
    data_dfs['L1'] = rep_dfs['L1']/data_dfs['N_cell']
    data_dfs['L2'] = rep_dfs['L2']/data_dfs['N_cell']
        
    data_dfs['B1'] = rep_dfs['B1']/data_dfs['N_cell']
    data_dfs['B2'] = rep_dfs['B2']/data_dfs['N_cell']
    data_dfs['T'] = rep_dfs['T']/data_dfs['N'] # Trunk is normalized with non-NAN cells
    data_dfs['S'] = rep_dfs['S']/data_dfs['N'] # Sholl_1 is normalized with non-NAN cells

    
    # save data
    print('saving...', DATA)
    with open(f'../{DATA}/morphology_data/rep_dfs.pkl', 'wb') as handle:
        pickle.dump(rep_dfs, handle)
    with open(f'../{DATA}/morphology_data/data_dfs.pkl', 'wb') as handle:
        pickle.dump(data_dfs, handle)
```

```python
# compile into one data. 
data = {}
names = ['D10_STD1','D10_-AO1','D10_-BNF1','D10_STD2',
        'D21_STD1','D21_-AO1','D21_-BNF1','D21_-AO2','D21_-BNF2',
        'D28_STD1','D28_-AO1','D28_-BNF1','D28_STD2','D28_-AO2','D28_-BNF2']
for i in range(len(DATA_DIRs)):
    name = names[i]
    DATA = DATA_DIRs[i]
    # load per sample
    with open(f'../{DATA}/morphology_data/data_dfs.pkl', 'rb') as handle:
        data[name] = pickle.load(handle)
```

```python
data['D10_STD1']['N_cell']
```

```python
# concatenate all features by gene
keys = ['N_cell','L','L1','L2','B','B1','B2','T','S','N']

cols = data['D10_STD1']['N_cell'].columns.tolist()
cols.remove('non-target')
new_cols = ['non-target'] + cols
print(new_cols)

# concatenate all features by gene
dfs = pd.DataFrame()
for name in names: # sample names
    for feat in keys:
        df_to_add = data[name][feat]
        df_to_add['Feature'] = feat
        df_to_add['Sample'] = name
        
        dfs=pd.concat([dfs,df_to_add], axis=0)
cols = dfs.columns.tolist()
cols.remove('non-target')
new_cols = ['non-target']+cols
dfs = dfs[new_cols]

# remove nan row
dfs = dfs[~dfs.isnull().any(axis=1)]

# create long df 
dfs['Field'] = dfs.index
long_df = pd.DataFrame()
features=['N_Cells','L', 'L1', 'L2', 'B','B1', 'B2','T', 'S','N']
for feature in features:
    df_to_add = dfs[dfs.Feature==feature]
    df_to_add = pd.melt(df_to_add, id_vars = ['Sample','Feature','Field'], var_name = ['Gene'])
    long_df = pd.concat([long_df, df_to_add], axis=0)
long_df=long_df.reset_index()
long_df
```

```python
# remove GAS7 and DGKE - low cell number at later timepoints; 
long_df = long_df[~long_df.Gene.isin(['DGKE','GAS7'])]
print(set(long_df.Gene))
```

```python
# normalize w.r.t each media D10

# construct norm_map - this is to calculate SURVIVAL Features
norm_map_per_media = {'D10_-AO1': 'D10_-AO1',
            'D10_-BNF1': 'D10_-BNF1',
            'D10_STD1': 'D10_STD1',
            'D10_STD2': 'D10_STD2',
            'D21_-AO1': 'D10_-AO1',
            'D21_-AO2': 'D10_-AO1',
            'D21_-BNF1': 'D10_-BNF1',
            'D21_-BNF2': 'D10_-BNF1',
            'D21_STD1': 'D10_STD1',
            'D28_-AO1': 'D10_-AO1',
            'D28_-AO2': 'D10_-AO1',
            'D28_-BNF1': 'D10_-BNF1',
            'D28_-BNF2': 'D10_-BNF1',
            'D28_STD1': 'D10_STD1',
            'D28_STD2': 'D10_STD2'}

for feat in set(long_df.Feature):
    for sample in set(long_df.Sample):
        df = long_df[(long_df.Feature == feat)&(long_df.Sample == sample)]
        ctrl_sample = norm_map_per_media[sample]
        ctrl_df = long_df[(long_df.Feature == feat)&(long_df.Sample == ctrl_sample)]
        print(sample, ctrl_sample)
        for gene in set(long_df.Gene): # for each gene
            # make sure to only include field info that's common to both dataset
            fields = df[df.Gene==gene].Field.values
            ctrl_inds = ctrl_df[(ctrl_df.Gene == gene)&(ctrl_df.Field.isin(fields))].index
            ctrl_fields = ctrl_df[(ctrl_df.Gene == gene)&(ctrl_df.Field.isin(fields))].Field.values
            inds = df[(df.Gene ==gene)].index
            long_df.loc[inds, 'normed_D10_value_per_media'] = np.divide(long_df.loc[inds, 'value'].values,np.mean(long_df.loc[ctrl_inds, 'value'].values))
long_df
```

```python
# update condition level metadata
# get a list of samples to combine.
cond_list = [x[:len(x)-1] for x in long_df.Sample.values]
long_df['Condition'] = cond_list

# Time, Media label
time_list = [x.split('_')[0] for x in long_df.Condition.values]
media_list = [x.split('_')[1] for x in long_df.Condition.values]
long_df['Time'] = time_list
long_df['Media'] = media_list
long_df
```

```python
# Flag Outliers
Conditions = set(long_df.Condition.values)
features = set(long_df.Feature.values)
for condition in Conditions:
    for feature in features:
        df = long_df[(long_df.Condition == condition)&(long_df.Feature==feature)]
        for gene in df.Gene.values:
            sub_df = df[df.Gene==gene]
            inds = sub_df.index
            q1,q3 = sub_df['value'].quantile(0.25), sub_df['value'].quantile(0.75)
            iqr=q3-q1
            flags = np.where((sub_df['value']<q1-1.5*iqr)|(sub_df['value']>q3+1.5*iqr), 1, 0)
            long_df.loc[inds, 'Outlier'] = flags # update long_df.
```

## Replicate-replicate correlation - Supplement

```python
# correlation - take the mean across all FOVs per replicates.
genes = list(set(long_df.Gene))
cols = ['L1_Rep1','L1_Rep2','B1_Rep1','B1_Rep2','S_Rep1','S_Rep2','N_Rep1','N_Rep2']
corr_dfs = {}
corr_dfs['D10_STD'] = pd.DataFrame(index = genes, columns = cols)
corr_dfs['D28_STD'] = pd.DataFrame(index = genes, columns = cols)
corr_dfs['D28_-AO'] = pd.DataFrame(index = genes, columns = cols)
corr_dfs['D28_-BNF'] = pd.DataFrame(index = genes, columns = cols)
features = ['L1','B1','S','N']
for key in corr_dfs.keys():
    df = corr_dfs[key]
    for feature in features:
        if feature != 'N':
            data_col = 'value'
        else:
            data_col = 'normed_D10_value_per_media'
        # outlier removed?
        data_df = long_df[(long_df.Condition == key)&(long_df.Feature == feature)&(long_df.Outlier != 1)]
        #data_df = long_df[(long_df.Condition == key)&(long_df.Feature == feature)]
        for gene in genes:
            data1 = data_df[(data_df.Gene == gene)&(data_df.Sample == f'{key}1')]
            data2 = data_df[(data_df.Gene == gene)&(data_df.Sample == f'{key}2')]
            df.loc[gene, f'{feature}_Rep1'] = np.mean(data1[data_col])
            df.loc[gene, f'{feature}_Rep2'] = np.mean(data2[data_col])
    # update
    corr_dfs[key] = df.copy()
```

```python
# D28 STD, D28 -AO and D28 -BNF
to_plot_list = ['D10_STD','D28_STD']
feature_annot_list = ['Neurite Length (pixels)', 'Number of Branches', 'Number of Trunks']
fig, axs = plt.subplots(2,3, figsize = (30,20))
for i in range(2):
    for j in range(3):
        key = to_plot_list[i]
        feature = features[j]
        sns.scatterplot(data = corr_dfs[key], x = f'{feature}_Rep1', y = f'{feature}_Rep2', ax=axs[i][j], s=75)
        x_vals = axs[i][j].get_xlim()
        y_vals = axs[i][j].get_ylim()
        min_val = min(x_vals[0], y_vals[0])
        max_val = max(x_vals[1], y_vals[1])
        axs[i][j].plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--')
        axs[i][j].spines['bottom'].set_color('black')
        axs[i][j].spines['left'].set_color('black')
        
        #ticks = axs[i][j].get_yticks()  # Use y-axis ticks as the base
        #axs[i][j].set_xticks(ticks)
        #axs[i][j].set_yticks(ticks)
        axs[i][j].tick_params(axis='both', which='major', labelsize=30)
        
        axs[i][j].set_xlabel(f'{feature_annot_list[j]}\nReplicate 1', fontsize = 40)
        axs[i][j].set_ylabel(f'{feature_annot_list[j]}\nReplicate 2', fontsize = 40)
        
        corr_coef, _ = pearsonr(corr_dfs[key][f'{feature}_Rep1'], corr_dfs[key][f'{feature}_Rep2'])
        axs[i][j].text(0.95, 0.05, f'Pearson r = {corr_coef:.2f}', transform=axs[i][j].transAxes,
                       fontsize=40, horizontalalignment='right', verticalalignment='bottom')
        
       
        # Customize spines
        axs[i][j].spines['top'].set_visible(False)
        axs[i][j].spines['right'].set_visible(False)
        axs[i][j].spines['bottom'].set_linewidth(5)
        axs[i][j].spines['left'].set_linewidth(5)
        axs[i][j].spines['bottom'].set_color('black')
        axs[i][j].spines['left'].set_color('black')
plt.tight_layout()
plt.savefig('../Figures_Fig5/Neurite_rep_rep_correlation_STD.png', format='png', dpi = 300)    
plt.savefig('../Figures_Fig5/Neurite_rep_rep_correlation_STD.eps', format='eps', dpi = 300)
```

## Plot example points - line graph related to Figure 4A

```python
##### USER INPUT ##### 
style_map = {'STD': 'solid',
            '-AO': 'dashed',
            '-BNF': 'dashdot'}
color_map = {'non-target': 'k',
            'MINK1': 'r',
            'PPP2R1A': 'b'}
marker_map = {'STD': 'o',
             '-AO': 's',
             '-BNF': '^'}
genes_to_plot = ['non-target','PPP2R1A','MINK1']

features = ['L1','B1','S']
media_to_plot = ['STD']
feature_annot_list = ['Neurite Length (pixels)', 'Number of Branches', 'Number of Trunks']

fig, axs = plt.subplots(1,3, figsize= (15,5))
for i in range(len(axs)):
    feat = features[i] # L1, B1, S
    # subset and aggregate by conditions.
    df_to_plot = long_df[(long_df.Feature == feat)&(long_df.Gene.isin(genes_to_plot))&(long_df.Outlier != 1)]
    agg_df = df_to_plot.groupby(['Gene','Condition'])['value'].agg(['mean','sem']).reset_index()
    # append media / time labels
    media_labels = [x.split('_')[1] for x in agg_df.Condition]
    time_labels = [x.split('_')[0] for x in agg_df.Condition]
    agg_df['Time'] = time_labels
    agg_df['Media'] = media_labels
    # for each gene and for each condition, plot 
    Times = ['D10','D21','D28']
    for gene in genes_to_plot:
        color = color_map[gene]
        for media in media_to_plot:
            style = style_map[media]
            marker = marker_map[media]
            df = agg_df[(agg_df.Gene == gene)&(agg_df.Time.isin(Times))&(agg_df.Media == media)].sort_values(by='Time')
            # plot time
            axs[i].errorbar(Times, df['mean'], yerr = df['sem'], marker = marker, linestyle = style, color = color,
                        capsize = 5, markersize = 2, linewidth = 2)
    handles = []
    for gene in genes_to_plot:
        color = color_map[gene]
        pp = Patch(color = color, label = gene)
        handles.append(pp)

    for media in media_to_plot:
        line = style_map[media]
        marker = marker_map[media]
        ll = Line2D([], [], color = 'black', label = media, linestyle = line, marker=marker,
                   markersize = 5, linewidth = 2)
        handles.append(ll)

    axs[i].set_xlabel('Time', fontsize=25)
    axs[i].set_ylabel(f'{feature_annot_list[i]}', fontsize=25) 
    axs[i].tick_params(axis='both', labelsize = 15)
    
    
    axs[i].set_facecolor('white')
    axs[i].spines['left'].set_color('black')
    axs[i].spines['bottom'].set_color('black')
    axs[i].spines['right'].set_color('white')
    axs[i].spines['top'].set_color('white')
    
    for spine in axs[i].spines.values():
        spine.set_linewidth(2.5)  # Adjust the thickness of the spines

    # Set whisker thickness
    #for whisker in b.artists:
    #    whisker.set_linewidth(2)
    
    # Set median line thickness
    for line in axs[i].lines:
        line.set_linewidth(2)
    
plt.tight_layout(pad=0.0, w_pad=1.0, h_pad=1.0)  # Adjust padding here for more spacing
plt.subplots_adjust(wspace=0.5)
plt.savefig('../Figures_Fig5/D28_STD_selected_genes.png', format='png', dpi = 300)    
plt.savefig('../Figures_Fig5/D28_STD_selected_genes.eps', format='eps', dpi = 300)
```

```python
# save results in a dataframe- initialize
# effect_size : mean(gene)/mean(non-target)
ttest_df = pd.DataFrame(columns = ['Feature','Condition','Gene', 'pval','p_adj','t_statistic','effect_size'])
features = ['N','L1','B1','S']
conditions = list(set(long_df.Condition))
genes = list(set(long_df.Gene))
genes.remove('non-target')

# perform t-test
i=0
for feat in features:
    for condition in conditions:
        sub_df = long_df[(long_df.Feature == feat)&(long_df.Outlier != 1)&(long_df.Condition == condition)]
        sub_df['normed_D10_value_per_media'] = pd.to_numeric(sub_df['normed_D10_value_per_media'], errors='coerce')
        sub_df['value'] = pd.to_numeric(sub_df['value'], errors='coerce')
        for gene in genes:
            if feat == 'N':
                array1 = sub_df[sub_df.Gene == gene].normed_D10_value_per_media.dropna()
                array2 = sub_df[sub_df.Gene == 'non-target'].normed_D10_value_per_media.dropna()
            else:
                array1 = sub_df[sub_df.Gene == gene].value.dropna()
                array2 = sub_df[sub_df.Gene == 'non-target'].value.dropna()
            
            t_statistic, p_value = ttest_ind(array1, array2)
            effect = np.mean(array1)/np.mean(array2)
            ttest_df.loc[i, ['Feature','Condition','Gene','pval','t_statistic','effect_size']] = [feat,
                                                                                                 condition,
                                                                                                 gene,
                                                                                                 p_value,
                                                                                                 t_statistic,
                                                                                                 effect]
            i+=1

# for each feature and condition combination, perform multiple test correction. 
for feat in features:
    for condition in conditions:
        sub_df = ttest_df[(ttest_df.Condition == condition)&(ttest_df.Feature == feat)]
        inds = sub_df.index
        # get index
        sub_df['p_adj'] = smm.multipletests(sub_df['pval'], method = 'fdr_bh')[1]
        ttest_df.loc[inds, 'p_adj'] = sub_df['p_adj']
ttest_df
```

```python
# create edges from AUC matrix
# df: pairwise comparisons showing AUC
def create_networkx_from_auc(df, thres, nodes):
    df = df.loc[nodes, nodes]
    df = df < thres # from AUC, smaller means "edge"
    node_labels = df.columns.tolist()
    df = df.astype('float').to_numpy()
    
    # Create a NetworkX graph from the adjacency matrix with node labels
    G = nx.from_numpy_matrix(df, create_using=nx.Graph())
    G = nx.relabel_nodes(G, {i: label for i, label in enumerate(node_labels)})
    
    return G

def jaccard_index_edges(G1, G2):
    # get jaccard index
    set1 = set(tuple(sorted(edge)) for edge in G1.edges())
    set2 = set(tuple(sorted(edge)) for edge in G2.edges())
    J = len(set1.intersection(set2))/len(set1.union(set2)) #
    return J

# shuffle and make edges
def permute_auc_dataframe(df, random_seed):
    #df=pairwise_clf_results_auc['D28_STD'].copy()
    vals = df.to_numpy()
    nodes_to_shuffle = df.columns.tolist()
    random.seed(random_seed)
    random.shuffle(nodes_to_shuffle)
    shuffled_df = pd.DataFrame(index = nodes_to_shuffle, columns = nodes_to_shuffle, data = vals)
    return shuffled_df

def degree_preserving_ppi_shuffling(G_in, rand_seed):
    deg_seq = [d for n,d in G_in.degree()] # construct degree info
    mapping = dict(zip(range(len(G_in.nodes())), G_in.nodes()))
    G_rand = nx.random_degree_sequence_graph(deg_seq, seed=rand_seed)
    G_rand = nx.relabel_nodes(G_rand, mapping)
    return G_rand

## Create neurite-similarity graph


def create_adj_matrix_from_arborization(ttest_df, feat, condition, permute, rand_seed):
    # Filter dataframe based on the condition and feature
    df = ttest_df[(ttest_df.Condition == condition) & (ttest_df.Feature == feat)].copy()
    
    # Permute genes if requested
    if permute:
        nodes_to_shuffle = df.Gene.tolist()
        random.seed(rand_seed)
        random.shuffle(nodes_to_shuffle)
        df['Gene'] = nodes_to_shuffle
    
    # Extract unique genes and their t_statistics
    genes = df.Gene.unique()
    t_stats = df.set_index('Gene').loc[genes, 't_statistic'].values
    
    # Compute outer product of t_statistics to identify ++ or --
    product_matrix = np.outer(t_stats, t_stats)
    
    # Create adjacency matrix: 1 if product > 0, otherwise 0
    adj_matrix = (product_matrix > 0).astype(float)
    
    # Convert to a DataFrame for compatibility
    L_adj_matrix = pd.DataFrame(adj_matrix, index=genes, columns=genes)
    
    return L_adj_matrix
```
